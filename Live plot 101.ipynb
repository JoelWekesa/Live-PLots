```python
import pandas as pd
import numpy as np
import time
import random
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
x = 0
y = 2
z = 1
Y = []
X = []
Z = []
while True: 
    try:
        x += 1
        y += random.randint(-5, 6)
        z += random.randint(-4, 7)
        X.append(x)
        Y.append(y)
        Z.append(z)
        ax.cla()
        ax.plot(X, Y, linewidth = 1, linestyle = ":", color = 'red')
        ax.plot(X, Z, linewidth = 1, color = 'green')
        fig.canvas.draw()
    except KeyboardInterrupt:
        print("You've decided to interrupt the kernel. Process will stop.")
        break
time.sleep(1)
```
